```python
customer_df.columns = customer_df.columns.str.lower()
```

```python
from datetime import datetime
customer_df['effective_to_date'] = pd.to_datetime(customer_df['effective_to_date'])
customer_df.effective_to_date
```

```python
num = customer_df[['customer_lifetime_value', 'income', 'monthly_premium_auto', 'months_since_last_claim', 'months_since_policy_inception', 'number_of_open_complaints', 'number_of_policies', 'total_claim_amount']]
cat = customer_df[['state', 'response', 'coverage', 'education', 'effective_to_date', 'employmentstatus', 'gender', 'location_code', 'marital_status', 'policy_type', 'policy', 'renew_offer_type', 'sales_channel', 'vehicle_class', 'vehicle_size']]
```

```python
num.corr()
# except for the corr btw total_claim_amount and monthly_premium_auto, the overall corr btw numerical variables is low
```

```python
for x in ['customer_lifetime_value', 'income', 'monthly_premium_auto', 'total_claim_amount']:
    plt.figure(figsize=(10,5))
    plt.xlim(0,2000)
    plt.xlabel(x)
    sns.kdeplot(customer_df[x],shade=True)
    plt.show()
```

```python
# the distributions of continuous variable are quite skewed
```

```python
for each in cat:
    plt.figure(figsize=(12,8))
    sns.countplot(x = each, data=customer_df)
```

```python
# customer_id doesn't contain any useful info or carry categorical meaning
```

```python
for one in ['customer_lifetime_value', 'income', 'monthly_premium_auto', 'total_claim_amount']:
     sns.boxplot(x=one, data=customer_df)
     plt.show()
```
